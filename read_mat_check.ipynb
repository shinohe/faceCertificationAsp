```python
# check_mat

%matplotlib inline
from scipy.io import loadmat
from logging import getLogger, StreamHandler, DEBUG
import matplotlib.pyplot as plt
import numpy as np
import cv2

logger = getLogger(__name__)
handler = StreamHandler()
handler.setLevel(DEBUG)
logger.setLevel(DEBUG)
logger.addHandler(handler)
logger.propagate = False


imdb_wiki_marge = loadmat("imdbface/imdb_wiki_marge.mat")
```

```python
# displaying all image data is slow because image data is large. recomend to limit
roopCount = 50
#roopCount = len(imdb_wiki_marge["age"][0])

fig = plt.figure(figsize=(15, roopCount*2))
for i in range(roopCount):
    fig.add_subplot(roopCount/2+1,5,i+1)
    npData = np.array(imdb_wiki_marge["image"][i])
    age = imdb_wiki_marge["age"][0][i]
    gender = "M" if imdb_wiki_marge["gender"][0][i] == 1 else "F"
    npData = cv2.cvtColor(npData, cv2.COLOR_BGR2RGB)
    plt.imshow(npData)
    plt.title("%s.%s" % (age, gender))
```
